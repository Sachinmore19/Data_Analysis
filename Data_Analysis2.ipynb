# 4th dataset


```python
cre=pd.read_csv("D:/python/Credit.csv")
cre.head(2)
```

```python
cre['Income'].sort_values()
```

```python
q =[]
for i in cre['Income']:
    if(i <=50):
        q.append('Poor')
    elif(i <=100):
        q.append('Middle')
    elif(i <=150):
        q.append('Upper_Middle')
    else:
        q.append('Rich_affluent')
```

```python
cre['class']=q
```
